```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
direct_sorted = pd.read_csv('sorted.csv')
back_sorted = pd.read_csv('back_sorted.csv')
not_sorted = pd.read_csv('not_sorted.csv')
```

```python
sns.set_style("darkgrid")
```

```python
sns.lineplot(x='n', y='mysort', data=direct_sorted)
sns.lineplot(x='n', y='mysort', data=not_sorted)
sns.lineplot(x='n', y='mysort', data=back_sorted)

plt.title('mysort')
plt.xlabel('number of elements')
plt.ylabel('time')

plt.legend(labels=['direct sorted array', 'not sorted array', 'back sorted array'])

plt.show()
```

```python
sns.lineplot(x='n', y='mysort', data=direct_sorted)
sns.lineplot(x='n', y='qsort', data=direct_sorted)

plt.title('direct sorted array')
plt.xlabel('number of elements')
plt.ylabel('time')

plt.legend(labels=['mysort', 'qsort'])

plt.show()
direct_sorted
```

```python
sns.lineplot(x='n', y='mysort', data=back_sorted)
sns.lineplot(x='n', y='qsort', data=back_sorted)

plt.title('back sorted array')
plt.xlabel('number of elements')
plt.ylabel('time')

plt.legend(labels=['mysort', 'qsort'])

plt.show()
back_sorted
```

```python
sns.lineplot(x='n', y='mysort', data=not_sorted)
sns.lineplot(x='n', y='qsort', data=not_sorted)

plt.title('not sorted array')
plt.xlabel('number of elements')
plt.ylabel('time')

plt.legend(labels=['mysort', 'qsort'])

plt.show()
not_sorted
```
